Now that we have overall averages set we will continue with Monthly, Daily and Hourly ones

```python
# Add hourly, daily, and monthly average columns for SO2
air_data_df['Hourly_average_SO2'] = air_data_df.groupby(air_data_df['DateTime'].dt.hour)['SO2'].transform('mean')
air_data_df['Daily_average_SO2'] = air_data_df.groupby(air_data_df['DateTime'].dt.date)['SO2'].transform('mean')
air_data_df['Monthly_average_SO2'] = air_data_df.groupby(air_data_df['DateTime'].dt.month)['SO2'].transform('mean')

# Let's see if it worked and we can use it 

display(air_data_df)
```

Since the first and the last averages differed we can use this code
